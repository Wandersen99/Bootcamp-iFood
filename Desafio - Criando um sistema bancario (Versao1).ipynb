# Desafio: Criando um sistema bancário com python

## Objetivo Geral: Criar um sistema bancário com as operações: sacar, depositar e visualizar extrato.

## Regras:
* **Operação de Depósito:** Deve ser possível depositar valores positivos para a minha conta bancária. Todos os depósitos devem ser armazenados em uma variável e exibidos na operação de extrato.


* **Operação de Saque:** O sistema deve permitir realizar 3 saques diários com limite máximo de R$ 500,00 por saque. Caso o usuário não tenha saldo em conta, o sistema deve exibir uma mensagem informando que não será possível sacar o dinheiro por falta de saldo.


* **Operação de Extrato:**Essa operação deve listar todos os depósitos e saques realizados na conta. No fim da listagem deve ser exibido o saldo atual da conta. Se o extrato estiver em branco, exibir a mensagem: "Não foram realizadas movimentações."

```python
import time 

def exibir_menu(menu):
    menu = f"""\n\tCONTA BANCÁRIA
    {linha()}
    Cliente: {nome}
    Saldo Atual: R$ {saldo:,.2f}
    {linha()}
    Aperte a tecla correspondente
    a operação que deseja realizar:

    [d] Depositar
    [s] Sacar
    [e] Extrato
    [q] Sair

    => """
    return menu


def linha():
    linha = '*'
    return linha * 32


nome = input(f'Olá, digite o seu nome:\n').strip().casefold().title()

extrato = """\n######## EXTRATO DA CONTA ########\n"""
extrato_inicial = extrato
limite_saque = 500
num_saques = 0
MAX_SAQUES_DIARIO = 3
lista_operacoes = ['d','s','e','q']

while True:
    hora_atual = time.ctime()
    day = time.gmtime().tm_mday
    mes = time.gmtime().tm_mon
    ano = time.gmtime().tm_year
    opcao = input(exibir_menu(menu)).strip().casefold()
    
    if opcao in lista_operacoes:
        
        if opcao == 'q': 
            break
        
        elif opcao == 'd':
            deposito = float(input('Insira o valor que deseja Depositar: R$'))
            linha()
                             
            if deposito <= 0:
                print('\n*ERROR: O VALOR DO DEPÓSITO DEVE SER MAIOR QUE R$ 0.00\nTente novamente!')
                continue
                
            else:
                print(f'OPERAÇÃO REALIZADA COM SUCESSO!\n{linha()}')
                extrato += f'''\nData:{day}/{mes}/{ano} Hora:{hora_atual[10:-5]}\n->DEPÓSITO: R${deposito:,.2f}\n'''
                saldo += deposito
                menu = exibir_menu(menu)
                
        elif opcao == 's':
            saque = float(input('Insira o valor que deseja Sacar: R$'))
            linha()
            
            if (num_saques < MAX_SAQUES_DIARIO) and ((saque <= limite_saque) and (saque > 0)) and (saldo >= saque):
                num_saques += 1
                saldo -= saque
                print(f'OPERAÇÃO REALIZADA COM SUCESSO!\n{linha()}')
                extrato += f'''\nData:{day}/{mes}/{ano} Hora:{hora_atual[10:-5]}\n->SAQUE: R${saque:,.2f}\n'''
                menu = exibir_menu(menu)
            
            else:
                
                print(f'''\n*OPERAÇÃO INVÁLIDA. 1 ou + motivos ocorreram:
            -> Verifique se o valor do saque não excedeu o valor do saldo. Saldo:R$ {saldo}|Saque:R$ {saque}
            -> Limite diário de saques = 3. | Saques Realizados hoje = {num_saques};
            -> Limite de saque por operação = R$ 500.00 | Tentativa de saque registrada: R$ {saque}
            -> Saque deve ser maior que R$ 0.00 | Valores negativos não serão processados pelo sistema!
                        ''')
                linha()
                continue
                
                 
        elif opcao == 'e':
            if extrato == extrato_inicial:
                extrato_inicial += f'-> Não foram realizadas movimentações.\n\nSaldo Atual: R$ {saldo:,.2f}'
                print(extrato_inicial)
                
            else:
                print(extrato)
                print(f'\nSALDO ATUAL: R$ {saldo:,.2f}\n')
                print("##################################")
                                   
    else:
        print(f'\n*ERROR: OPERAÇÃO INVÁLIDA. Digite novamente a opção CORRETA!\n')
        linha()
        continue

print(f'\nObrigado por usar nossos serviços. Até logo, {nome}!')
```
